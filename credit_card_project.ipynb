```python
# converting to array 
x= np.array(credit_df_x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
y_test
```

```python
# since we have to transform an array into z score we are using stat 
from scipy import stats
```

```python
y_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
#importing support vector classifier
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train_scaled,y_train)
```

```python
# determining accuracy of the data set
print(" accuracy of the training data: {:.3f}".format(
    svc.score(x_train_scaled,y_train)))
print("accuracy of the test data: {:.3f}".format(
    svc.score(x_test_scaled,y_test)))
```

```python
# checking accuracy again with help of c=1000 technique
svc=SVC(C=1000)
svc.fit(x_train_scaled,y_train)
m=svc.predict(x_test_scaled)
print(m)
print(" accuracy of the training data: {:.3f}".format(
    svc.score(x_train_scaled,y_train)))
print("accuracy of the test data: {:.3f}".format(
    svc.score(x_test_scaled,y_test)))
```

```python
#using confusion matrix 
from sklearn import metrics
print(metrics.confusion_matrix(y_test,m))
```

```python
#recall and precision
#Precision can be seen as a measure of quality, and recall as a measure of quantity.
#recall formula for 0 is:(TP/TP+FN))::(85298/(85298+10)):false negative, actually 0 but treated as 1 : 99.99%
#recall formula for 1 is:(TP/TP+FN))::(54/(54+81)) :false negative,actually 1 but treated as 0: 40.0%
#precision formula for 0 is:(TP/TP+FP))::(85298/(85298+81)) : false positive, actually 0 but treated as 1: 99.9%
#precision formula for 1 is:(TP/TP+FP)):: (54/(54+10)) :false positive,actually 1 but treated as 0: 84.3%
```
